**Q.1. What is Simple Linear Regression?**
  - Simple linear regression (SLR) is a statistical method used to model the relationship between two continuous variables. One variable, known as the independent variable or predictor (conventionally denoted as x), is used to predict the value of the other, the dependent variable or response (conventionally denoted as y). The term "simple" indicates that there is only one predictor variable. The model seeks to find a straight line that best fits the data points on a scatter plot, minimizing the difference between the predicted values and the actual observed values.

  The Linear Regression Equation
  The formula for simple linear regression is represented by a linear equation:
  y = a + bx
  * y is the predicted value of the dependent variable.
  * a is the y-intercept, the predicted value of y when x is 0.
  * b is the slope of the regression line, representing the change in y for every one-unit increase in x.
  * x is the independent variable.

  How It Works
  The most common method for determining the "best-fit" line is the Ordinary Least Squares (OLS) method. This approach calculates the values of the slope (b) and the y-intercept (a) that minimize the sum of the squared vertical distances from each data point to the regression line. By doing so, the model finds the line that is closest to all the data points collectively. This line can then be used to predict the value of the dependent variable for a new, unobserved value of the independent variable.


**Q.2. What are the key assumptions of Simple Linear Regression?**
  - Simple linear regression relies on several key assumptions about the data and the model's errors (residuals). When these assumptions are met, the regression model is more reliable, and the statistical tests performed on it are valid. Violations can lead to inaccurate predictions and misleading conclusions.

  1. Linearity
  The relationship between the independent variable (x) and the dependent variable (y) must be linear. This means that the data points should approximate a straight line. If the true relationship is curved or non-linear, a simple linear model will not accurately capture the trend.

  2. Independence of Errors
  The errors (or residuals), which are the differences between the observed values and the values predicted by the model, should be independent of each other. This is particularly important for time-series data, where the error at one point in time might be correlated with the error at the next. This issue is known as autocorrelation.

  3. Homoscedasticity
  The variance of the errors should be constant across all levels of the independent variable. This is known as homoscedasticity. If the spread of the residuals increases or decreases as the value of the independent variable changes, it's a violation called heteroscedasticity. This can make predictions less reliable for certain ranges of the independent variable.

  4. Normality of Errors
  The errors should be normally distributed. While this assumption is less critical for the accuracy of predictions, it's essential for the validity of statistical significance tests (like t-tests and F-tests) and for calculating confidence and prediction intervals. You can check for this assumption by plotting a histogram of the residuals.


**Q.3. What is heteroscedasticity, and why is it important to address in regression models?**
  - Heteroscedasticity is a violation of a key assumption in linear regression where the variance of the residuals (errors) is not constant across all levels of the independent variable(s). Visually, this often appears in a residual plot as a "fan" or "cone" shape, where the spread of the residuals either increases or decreases as the predicted values (or the independent variable) change. The term itself comes from the Greek "hetero," meaning "different," and "skedasis," meaning "scattering."
  
  Why It's a Problem
  While heteroscedasticity does not cause the regression coefficient estimates to be biased, it does lead to several significant issues that can undermine the validity of your model:
    
    * Incorrect Standard Errors: The most critical problem is that heteroscedasticity causes the standard errors of the regression coefficients to be biased. They are often underestimated, meaning that the confidence intervals for your coefficients will be too narrow.
    
    * Misleading Statistical Tests: Because the standard errors are incorrect, the t-tests and F-tests used to determine the statistical significance of the coefficients and the overall model are also invalid. This can lead to a higher chance of a Type I error (incorrectly rejecting a true null hypothesis) and drawing misleading conclusions about the relationships between variables.
    
    * Inefficient Estimates: The Ordinary Least Squares (OLS) method, which is commonly used to fit regression models, is no longer the most efficient estimator when heteroscedasticity is present. This means there are other methods that could produce a more precise and reliable model from the same data.

  How to Address It
  To address heteroscedasticity, you can:
    * Transform Variables: A common approach is to apply a mathematical transformation to the dependent variable, such as a logarithmic or square root transformation. This can help to stabilize the variance and make the relationship more linear.

    * Use a Different Model: Consider using a different type of regression model, such as a Generalized Linear Model, that is designed to handle non-constant variance.

    * Use Robust Standard Errors: A more modern and often preferred solution is to use robust standard errors (also known as "White's" or "Huber-White" standard errors). These are adjusted standard errors that are valid even in the presence of heteroscedasticity, allowing you to use the same OLS coefficient estimates while still performing valid statistical inference.


**Q.4. What is Multiple Linear Regression?**
  - Multiple linear regression is a statistical technique used to model the relationship between one dependent variable and two or more independent variables. It is an extension of simple linear regression, which only uses one independent variable. The goal is to determine the best-fit linear equation that can predict the value of the dependent variable based on the values of multiple independent variables.

  The Multiple Linear Regression Equation
  The equation for a multiple linear regression model is:
  y = a + b_1x_1 + b_2x_2 + ...[span_3](end_span) + b_nx_n + \epsilon
    * y is the predicted value of the dependent variable.
    * a is the y-intercept, the predicted value of y when all independent variables are zero.
    * b_1, b_2, ..., b_n are the regression coefficients (slopes) for each independent variable. Each coefficient represents the change in y for a one-unit increase in its corresponding x variable, while holding all other x variables constant.
    * x_1, x_2, ..., x_n are the multiple independent variables.
    * \epsilon is the error term or residual.

  Key Differences from Simple Linear Regression
    * Number of Variables: The most fundamental difference is the number of independent variables. Simple linear regression has one, while multiple linear regression has two or more.

    * Dimensionality: A simple linear regression can be visualized as a two-dimensional line on a scatter plot. A multiple linear regression with two independent variables can be visualized as a three-dimensional plane, and with more variables, it becomes a "hyperplane" that is difficult to visualize but still follows the same principles.

    * Complexity and Interpretation: Multiple linear regression is more complex because it allows you to analyze the combined effect of multiple factors on a single outcome. It also helps to determine the relative contribution and importance of each independent variable in the model.


**Q.5. What is polynomial regression, and how does it differ from linear regression?**
  - Polynomial regression is a form of regression analysis in which the relationship between the independent variable (x) and the dependent variable (y) is modeled as an n^{th} degree polynomial. This technique is used when the data exhibits a non-linear relationship that a straight line cannot capture.

  The general equation for a polynomial regression model is:
  y = b_0 + b_1x + b_2x^2 + \ldots + b_nx^n
    * y is the predicted value of the dependent variable.
    * b_0, b_1, \ldots, b_n are the regression coefficients.
    * x is the independent variable.
    * n is the degree of the polynomial, which determines the complexity of the curve.

  How It Differs from Linear Regression
  | Feature | Linear Regression | Polynomial Regression | |---|---|---|
  | Relationship | Models a linear relationship between variables. The model fits a straight line to the data. | Models a non-linear relationship. The model fits a curve to the data. |
  | Equation | y = a + bx | y = b_0 + b_1x + \ldots + b_nx^n |
  | Flexibility | Less flexible. It's a simple model that works best when the relationship is clearly linear. | More flexible. It can capture complex curves and patterns in the data. |
  | Overfitting Risk | Lower risk of overfitting. | Higher risk of overfitting, especially with higher-degree polynomials. This is because a high-degree polynomial can fit the "noise" in the data rather than the underlying pattern. |

  Despite these differences, polynomial regression is technically considered a special case of multiple linear regression. This is because the equation can be seen as a linear combination of transformed variables (i.e., x, x^2, x^3, etc.). The model is linear in the parameters (the b coefficients), even though it fits a non-linear curve to the data.

**Q.6. Implement a Python program to fit a Simple Linear Regression model to the following sample data:**

● X = [1, 2, 3, 4, 5]

● Y = [2.1, 4.3, 6.1, 7.9, 10.2]

**Plot the regression line over the data points.**

**(Include your Python code and output in the code box below.)**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# The provided sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create a LinearRegression model and fit it to the data
model = LinearRegression()
model.fit(X, Y)

# Get the coefficients (slope and intercept)
slope = model.coef_[0]
intercept = model.intercept_

# Print the results
print(f"Intercept (a): {intercept:.4f}")
print(f"Slope (b): {slope:.4f}")

# Make predictions using the fitted model
Y_pred = model.predict(X)

# Plot the original data points
plt.scatter(X, Y, color='blue', label='Actual Data Points')

# Plot the regression line
plt.plot(X, Y_pred, color='red', linewidth=2, label='Regression Line')

# Add labels and a title
plt.title('Simple Linear Regression')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.legend()
plt.grid(True)
plt.show()
```

**Q.7. Fit a Multiple Linear Regression model on this sample data:**

● Area = [1200, 1500, 1800, 2000]

● Rooms = [2, 3, 3, 4]

● Price = [250000, 300000, 320000, 370000]

**Check for multicollinearity using VIF and report the results.**

**(Include your Python code and output in the code box below.)**

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 1. Create a DataFrame from the sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

# 2. Define the independent variables (X) and the dependent variable (y)
X = df[['Area', 'Rooms']]
y = df['Price']

# 3. Add a constant to the independent variables for the statsmodels OLS model
X = sm.add_constant(X)

# 4. Fit the Multiple Linear Regression model
model = sm.OLS(y, X).fit()

# 5. Print the regression summary to see the coefficients, R-squared, etc.
print("--- Multiple Linear Regression Model Summary ---")
print(model.summary())
print("\n" + "="*50 + "\n")

# 6. Calculate VIF for each independent variable
print("--- VIF Scores (Multicollinearity Check) ---")
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif_data)
```

**Q.8. Implement polynomial regression on the following data:**

● X = [1, 2, 3, 4, 5]

● Y = [2.2, 4.8, 7.5, 11.2, 14.7]

**Fit a 2nd-degree polynomial and plot the resulting curve.**

**(Include your Python code and output in the code box below.)**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# The provided sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create a 2nd-degree polynomial feature transformation
polynomial_features = PolynomialFeatures(degree=2)
X_poly = polynomial_features.fit_transform(X)

# Create and fit the linear regression model on the polynomial features
model = LinearRegression()
model.fit(X_poly, Y)

# Get the coefficients
intercept = model.intercept_
coefficients = model.coef_
print(f"Intercept: {intercept:.4f}")
print(f"Coefficients: {coefficients[1:]}")

# Create a smooth line for plotting the curve
X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
X_plot_poly = polynomial_features.fit_transform(X_plot)
Y_plot = model.predict(X_plot_poly)

# Plot the original data points
plt.scatter(X, Y, color='blue', label='Actual Data Points')

# Plot the polynomial regression curve
plt.plot(X_plot, Y_plot, color='red', linewidth=2, label='2nd-Degree Polynomial Fit')

# Add labels and a title
plt.title('Polynomial Regression (2nd Degree)')
plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.legend()
plt.grid(True)
plt.savefig("polynomial_regression_plot.png")
```

**Q.9. Create a residuals plot for a regression model trained on this data:**

● X = [10, 20, 30, 40, 50]

● Y = [15, 35, 40, 50, 65]

**Assess heteroscedasticity by examining the spread of residuals.**

**(Include your Python code and output in the code box below.)**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# The provided sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Create and fit a simple linear regression model
model = LinearRegression()
model.fit(X, Y)

# Get the predicted values
Y_pred = model.predict(X)

# Calculate the residuals
residuals = Y - Y_pred

# Create the residuals plot
plt.figure(figsize=(8, 6))
plt.scatter(Y_pred, residuals, color='blue', alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
plt.title('Residuals Plot')
plt.xlabel('Predicted Y values')
plt.ylabel('Residuals (Actual Y - Predicted Y)')
plt.grid(True)
plt.show()

# Assess heteroscedasticity
print("--- Assessment of Heteroscedasticity ---")
print("Heteroscedasticity is indicated by a non-random, funnel, or cone-shaped pattern in the residuals.")
print("The residuals in this plot appear to be randomly scattered around the zero line with no clear pattern.")
print("This suggests that the variance of the residuals is constant across all predicted values,")
print("indicating that the assumption of homoscedasticity is likely met.")
```

**Q.10. Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model.**
  - As a data scientist, here are the steps I would take to address heteroscedasticity and multicollinearity in the house price prediction model.

  Addressing Heteroscedasticity
  Heteroscedasticity, where the variance of the errors is not constant, can lead to biased standard errors, invalid t-tests, and unreliable confidence intervals. I would address this in the following ways:

    * Transform the Dependent Variable: The first and most common step is to apply a logarithmic or square root transformation to the house prices. For example, using log(Price) can compress the range of the dependent variable and often stabilize the error variance.

    * Use Robust Standard Errors: I would use robust standard errors (also known as Huber-White standard errors). These standard errors are calculated in a way that remains valid even when the assumption of homoscedasticity is violated. Many statistical software packages and libraries (like statsmodels in Python) have a built-in option for this.

    * Weighted Least Squares (WLS): If I can identify the specific pattern of the heteroscedasticity, I could use WLS. This method assigns lower weights to observations with higher variance and higher weights to observations with lower variance, effectively down-weighting the noisy data points.

  Addressing Multicollinearity
  Multicollinearity occurs when independent variables are highly correlated with each other, making it difficult to determine the individual effect of each variable. This can lead to inflated standard errors and unstable regression coefficients. Here's how I would tackle it:
    
    * Remove Correlated Variables: The simplest approach is to identify the pairs of highly correlated variables using a correlation matrix or VIF (Variance Inflation Factor) scores. If two variables like "number of bedrooms" and "number of rooms" are highly correlated, I would consider keeping only one of them. I'd likely keep the one that is more interpretable or has a stronger individual correlation with the price.

    * Combine Variables: I would create a new feature by combining the correlated ones. For instance, instead of using separate variables for "area" and "number of rooms," I might create a "room density" variable (e.g., Area / Rooms). This can reduce multicollinearity while preserving the information.

    * Principal Component Analysis (PCA): For more complex cases with multiple correlated variables, I would use Principal Component Analysis. PCA transforms the correlated features into a new set of uncorrelated variables called principal components. These new components can then be used as predictors in the regression model.

    * Regularization Techniques: I would consider using regularization methods like Ridge Regression or Lasso Regression. These techniques add a penalty term to the regression equation that shrinks the coefficients. Ridge regression is particularly effective at handling multicollinearity by reducing the variance of the coefficient estimates.
